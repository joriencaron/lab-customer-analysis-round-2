```python
import pandas as pd
import numpy as np
```

```python
#Show the dataframe shape.
df_marketing = pd.read_csv('marketing_customer_analysis.csv')
df_marketing
```

```python
df_marketing = df_marketing.drop(['Unnamed: 0'], axis=1)
df_marketing
```

```python
cols = []
for c in df_marketing.columns:
    cols.append(c.lower())
df_marketing.columns = cols

df_marketing
```

```python
df_marketing.dtypes
```

```python
df_marketing.isna()
```

```python
df_marketing.isna().sum()
```

```python
df_marketing = df_marketing.drop(['vehicle type'], axis=1)
```

```python
mean_months_claim = np.mean(df_marketing['months since last claim'])
df_marketing['months since last claim'] = df_marketing['months since last claim'].fillna(mean_months_claim)

mean_open_complaint = np.mean(df_marketing['number of open complaints'])
df_marketing['number of open complaints'] = df_marketing['number of open complaints'].fillna(mean_open_complaint)
df_marketing
```

```python
df_marketing['state'].value_counts()
df_marketing['response'].value_counts()
df_marketing['vehicle class'].value_counts()
df_marketing['vehicle size'].value_counts()
```

```python
df_marketing['state'] = df_marketing['state'].fillna('California')
df_marketing['response'] = df_marketing['response'].fillna('No')
df_marketing['vehicle class'] = df_marketing['vehicle class'].fillna('Four-Door Car')
df_marketing['vehicle size'] = df_marketing['vehicle size'].fillna('Medsize')
```

```python
df_marketing['effective to date'] = pd.to_datetime(df_marketing['effective to date'], errors='coerce')
```

```python
df_marketing['effective to date'].dt.month
df_marketing['months'] = df_marketing['effective to date'].dt.month

df_marketing
```

```python
df_marketing[df_marketing['months']<4]
```

```python
def transform_date(dataframe, column, new_column):
    dataframe[column] = pd.to_datetime(dataframe[column], errors='coerce')
    dataframe[new_column] = dataframe[column].dt.month

    return dataframe[dataframe[new_column]<4]

transform_date(df_marketing, 'effective to date', 'months')
```

```python
df_marketing.to_csv('marketing_customer_analysis-round2.csv')
```
